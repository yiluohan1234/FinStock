## 1.2 均线的趋势

如今，打开任意一款看盘软件，都能极其方便地调出均线。可是，要想更好地运用均线技术，首先就有必要真正理解均线的本质意义。

我们知道，**股票价格是由市场买卖决定的**。当股票的现价是 8.71 元时，就意味着有投资者以 8.71 元的价格买入，同时又有投资者以 8.71 元的价格卖出。因此，**股价是市场中一部分投资者的买入价，也是另一部分投资者的卖出价。**那么，**5 日均价就是市场 5 日来的平均成本价，也是市场 5 日来的平均销售价。**①

①严格来说，这个表述并不十分精确，因为均线统计的只是每日收盘价，并未统计当天运行的任一时刻的成交价。但作为应用于实战的技术分析，我们只能这样处理。投资者如果觉得有必要，可以将周期由日线图切换到 60 分钟线、30 分钟线、15 分钟线、5 分钟线，乃至 1 分钟线图。

当股价高于 5 日均价时，表明 5 日之内的买入者平均下来是盈利的。如果股价持续高于 5 日均价，就表明在一段时间内买入股票的投资者平均下来都是盈利的。这种获利效应会诱使更多的投资者进场买股票，从而进一步推高股价。在这种情况下，股价将总体维持在 5 日均线上方运行，而 5 日均线也必定是持续向右上方移动的，我们将这样的走势称为“上涨趋势”或“头趋势”。之后即便股价一度跌破 5 日均线，只要 5 日均线的运行方向是往上的，则我们仍然认定这属于多头趋势。在多头趋势中，投资者以买入和持有股票为主。

反之，当股价低于 5 日均价时，表明 5 日之内的买入者平均下来是亏提的。如果股价持续低于 5 日均价，就表明在一段时间内买入股票的投资者平均下来都是亏损的。这种沮丧的情绪会导致场内投资者倾向于卖出股票，从而进一步打压股价。在这种情况下，股价将总体维持在 5 日均线下方运行面 5 日均线也必定是持续向右下方移动的，我们将这样的走势称为“下跌趋势或“空头趋势”。之后即便股价一度站上 5 日均线，只要 5 日均线的运行方向还是向下的，则我们仍然认定这属于空头趋势。在空头趋势中，投资者以卖出和做空股票为主。

图 1.2-1 是日照港（600017）自 2015 年 5 月至 7 月的日线走势图：

```python
k_121 = KLineChart("600017", start_date='20150505', end_date='20150728')
k_121.plot(
    n=5,
    area=['V','BIAS'],
    width=1000,
    height=600,
    klines=['ma5'],
    is_notebook=True
)
```

可以看到，日照港 5 月 19 日的股价收盘报 6.01 元，突破 5 日均价 5.93 元，站上了 5 日均线。此后日照港股价一路震荡上扬，5 日均线也随之一路走高。6 月 18 日，日照港股价涨至最高点 11.68 元，5 日均线也随之涨至最高的 10.28 元。

在这一个月的时间里，由于 5 日均线始终是不断上涨的，所以我们将这一段走势视为多头趋势。其中，5 月 28 日和 29 日的收盘价虽然在 5 日均线下方，但由于它们并未扭转 5 日均线向上运行的方向，所以我们仍然认定这是多头趋势，投资者可以继续持股。果然，紧接着股价又重新站回了 5 日均线之上。

由此可见，我们判断趋势是多头还是空头，并不是看某一天，或某几天的股价涨跌，而是看均线的运行方向。当均线向上移动时，即便股价短期内有所下跌，之后也会顺着趋势的力量继续上涨，除非股价持续下跌，以至于扭转了均线的运行方向。这就是均线的趋势意义。

在 6 月 18 日创出 11.68 元的最高价后，日照港于 19 日大跌 9.74%，收报 9.91 元，已经跌破 5 日均线，5 日均线也由原来的上升趋势开始走平，仍在 10.28 元。此时尚不便断言多头趋势已经结束，但亦足以高度警惕。

23 日（紧接 19 日），日照港股价继续下跌，跌幅为 3.13%，收盘报 9.60 元，5 日均线则随之下移至 10.20 元。由于均线下行，已经初步显露出多头趋势结束的迹象，且股价近期涨幅过大，仅一个月时间就实现了翻倍，因此谨慎的投资者此时就可以考虑卖出了。

24 日，日照港股价出现反弹，涨幅为 4.79%，收盘报 10.06 元，5 日均线在 10.25 元。由于股价未能重新站回 5 日均线之上，故 23 日卖出的投资者此时不宜再买回，没有卖出的投资者同样可以考虑在本日卖出。

25 日，日照港股价再次大跌，跌幅为 6.86%，收盘报 9.37 元，5 日均线随之下移至 9.98 元。至此，5 日均线下移的趋势已经日益明显，此前未能卖出的持股者当考虑在此日卖出，甚至激进的投资者可考虑在本日融券做空。

此后，日照港股价一路下跌，最低跌至 7 月 8 日的 5.11 元，5 日均线随之下降至 6.18 元。其间虽偶有反弹，但 5 日均线自始至终都是保持向下动的，故而这段走势一直是空头趋势。站在 5 日均线的角度来说，投资者当以卖出股票为主，甚至融券做空，而不应该盲目抄底抢反弹。

7 月 9 日，日照港股价涨停，收盘报 5.86 元，但 5 日均线还是下移的在 5.92 元。因此，我们只能将这次涨停视为跌深之后的强劲反弹，就趋势而言，仍属于空头趋势。

10 日，日照港股价再次涨停，收盘报 6.45 元，并将 5 日均线由此前的下跌扭转为走平，在 5.93 元。这样，我们可以视作空头趋势即将结束，多头趋势或将启动。

13 日（紧接着 10 日），日照港股价第三度涨停，收盘报 7.10 元，并得5日均线明显拐头向上。此时，站在 5 日均线的角度而言，我们就可以确认多头趋势已经开始，可以考虑进场买股票了。①

①当股价跌到 7 月 8 日时，其实是绝佳的抄底良机，后文会详细介绍捕提此良机的方法。但仪凭现讨论的 5 日均线，是不宜进场抄底的，而只能保持观望态度。纯接照 5 日均线的趋势操作，抄间应该是在 13 日、即 5 日均线正式拐头向上时买入。可是如果得运用均线抵扣的技术，那么早 10 日便能推测出到 13 日的时候 5 日均线必定会拐头向上，这样就多赚了一个涨停板。如果懂得运乖离率的技术，则在 8 日或 9 日就可以知道跌幅过深之后必将会有强劲反弹，这样就又能多赚一个涨停板。这些技术都将在后文中一一介绍。

## 1.3 均线的参数

**均线是趋势的研判指标**，如果股价与均线呈现纠缠态势，时而位于均线之上，时而又位于均线之下，致使均线时而上行，时而下行，我们就将这的走势称为**“震荡走势”或“无趋势”**。

图 1.3-1 是东风汽车（600006）自 2018 年 6 月至 10 月的日线走势图：

```python
k_131 = KLineChart("600006", start_date='20180620', end_date='20181010')
k_131.plot(
    n=5,
    area=['V','BIAS'],
    width=1000,
    height=600,
    klines=['ma5'],
    is_notebook=True
)
```

可以看到，东风汽车的股价时而位于 5 日均线之上，时而位于 5 日均线之下，致使 5 日均线时而上行，时而下行。因此，站在 5 日均线的角度而言，这段走势就是震荡走势，是没有趋势的。

对于震荡走势而言，如果按照股价站上均线就买入持有，跌破均线就卖出做空的方法来操作，那就会导致来回犯错，亏损连连。以图 1.3-1 的东风汽车为例，当投资者因股价站上均线而买入时，没多久就不得不因股价跌破均线而卖出，且卖出价通常都会低于买入价。当投资者因股价跌破均线而卖出时，没多久就不得不因股价站上均线而买回，且买入价通常都会高于卖出价虽然每次的亏损金额都不算太大，可一旦来回次数多了，累计起来也不是个小数目，而且会加剧投资者内心的焦躁和挫败感。

图 1.3-2 是白云机场（600004）自 2019 年 1 月至 10 月的日线走势图：

```python
k_132 = KLineChart("600004", start_date='20190107', end_date='20191022')
k_132.plot(
    n=5,
    area=['V','BIAS'],
    width=1000,
    height=600,
    klines=['ma5'],
    is_notebook=True
)
```

可以看到，白云机场的股价也是时面位于 5 日均线之上、时面位于 5 日均线之下，致使 5 日均线时而上行，时面下行。因此，站在 5 日均线的角度而这段走势就是震荡走势，是没有趋势的。如果投资者是因股价站上 5 日就买入，跌破 5 日均线就卖出，那么看似股价翻了一倍还不止，可到头来但不会有多少盈利，甚至还可能会亏钱。

可间题在于、我们能清楚看到，自云机场的走势明明就是震荡上行品虽然上行途中也有回落，但大趋势肯定是向上，这是毫无疑问的，怎么它没有趋势呢？在不到九个月的时间内，白云机场的股价从最低 9.22 元最高 23.55 元，涨幅高达 155.42%，这难道不是靠着趋势的力量码！要都不算上涨造势，那得怎样才能算上涨趋势呢？

我们再来看图 1.3-3：

```python
k_133 = KLineChart("600004", start_date='20190107', end_date='20191022')
k_133.plot(
    n=60,
    area=['V','DKC'],
    width=1000,
    height=600,
    klines=['ma60'],
    is_notebook=True
)
```

我们在图中增加了一条 60 日均线。显然，白云机场的股价始终是位于 60 日均线之上，而 60 日均线也是一路上行的。因此，站在 60 日均线的角度而言，白云机场的股价确实是处于上涨趋势的。

站在 5 日均线的角度而言，白云机场基本上是无趋势方向的，此时若要顺势操作，就极易来回受伤。相对来说，高抛低吸反倒是更好的操作，即便不能全程赚满，但多少也能有些获利，且肯定不会亏钱。站在 60 日均线的角度而言，白云机场一直是处于上涨趋势，那么最好的操作就是顺势而为，在趋势结束之前坚定持股，以获得极为丰厚的收益。

通过这个案例就能明白，**在研判股价的运行趋势时，首先就得注意是站在怎样的时间维度**。因此，为了更好地通过均线研判趋势，我们通常不会仅凭一根均线，而是设置多条均线，以进行综合评价。

按照一般经验，我们会在股价的日线图中设置 5 日、10 日、20 日、60 日 120 日和 250 日六条均线。之所以这样设置参数，主要是因为在我国现行的交易制度下，5 日的是一周，10 日对应的是两周， 20 日对应的是一个月，60 日对应的是一季度，120 日对应的是半年，250 日对应的是一年。因此，我们习惯上 5 日均线称为周线，将 20 日均线称为月线，将 60 日均线称为季线，将 120 日均线称为半年线，将 250 日均线称为年线。

## 1.4 均线的性质

**均线经过对收盘价的移动平均，其走势不再像收盘价那样忽上忽下，而是显示出相当的稳定性，这也正是均线可以用来研判趋势的根本原因。**所谓趋势，指的是事物发展的运动方向。倘若运动方向杂乱无章，那自然也就谈不上趋势了。图 1.4-1 是中国国贸（600007）自 2016 年 1 月至 11 月的日线走势图：

```python
k_141 = KLineChart("600007", start_date='20160108', end_date='20161108')
k_141.plot(
    n=60,
    area=['V','DKC'],
    width=1000,
    height=600,
    klines=['ma60'],
    is_notebook=True
)
```

可以看到，在这十个月的时间里，中国国贸的日 K 线有阴有阳，有涨有跌。如果缺乏可以信赖的研判指标，则投资者极有可能会畏首畏尾，想买又怕一买就跌，想卖就怕一卖就涨。事实上，在这轮近乎翻倍的大行情里，绝大多数散户只能赚到 3%、5%，甚至还有很多人会亏钱。这就是因为他们的思维过多地受到忽上忽下的短期涨跌的影响，他们的情绪过多地受到嘈杂喧器的市场舆论的左右，而**无法对股价的中长期走势做出明确稳定的判断**。

如果我们能够妥善运用均线技术，就可以避免出现这种问题了，因为在图中绝大多数时间内，60 日均线都是稳步上行的。我们可以在 60 日均线上行时买入，至 60 日均线下行时再卖出。在此期间，只需安心持股即可，就不必再为“今天买了，明天要不要卖”或者“今天卖了，明天要不要买回来”而烦恼了。

4 月 6 日，中国国贸的 60 日均线在 13.54 元。7 日，其 60 日均线 13.51 元。8 日（周五），其 60 日均线在 13.51 元。11 日（周一），其 60 日均线在 13.52 元，60 日均线自此正式进入上行趋势。11 日的收盘价为 14.68 元，假设我们是以当日的收盘价买入。

10 月 20 日，中国国贸的 60 日均线在 20.88 元。21 日（周五），其 60 日均线在 20.89 元。24 日（周一），其 60 日均线在 20.89 元。25 日，其 60 日均线在 20.89 元。26 日，其 60 日均线在 20.89 元。27 日，其 60 日均线在 20.87 元，60 日均线自此正式进入下行趋势。27 日的收盘价为 18.65 元，假设我们是以当日的收盘价卖出，则收益率为 18.65 **÷** 14.68 - 1 = 27.04%

正是由于均线所具有的稳定性，我们才能够持有该股长达半年之久，获得三成左右的利润。然而，**凡事有一利就有一弊**，在**具备稳定性优点的同时均线也存在滞后性的缺点**。

依旧是以图 1.4-1 的中国国贸为例。可以看到，中国国贸的股价见底是在 1 月 14 日的 11.91 元，可是我们根据 60 日均线上行买入是在 4 月 11 日的 14.68 元。买入时间滞后了三个月，买入价格高出了 23.26%。中国国贸的股价见顶是在 8 月 15 日的 23.02 元，可是我们根据 60 日均线下行卖出是在 10 月 27 日的 18.65 元。卖出时间滞后了两个半月，卖出价格低 18.98%。

事实上，**稳定性和滞后性对于均线原本就是一体两面，为了获得其稳定性所带来的趋势研判价值，我们必须忍受其固有的滞后性这一缺点**。显然，**均线的周期越长就越稳定，同时也越滞后**。均线的周期越短就越不稳定，同时也越灵敏。如果我们将均线周期设为 1，那它事实上就是收盘价的连线了。

图 1.4-2 显示了中国国贸自 2016 年 1 月至 12 月间的 5 日、60 日、250 日均线走势：

```python
k_142 = KLineChart("600007", start_date='20160108', end_date='20161230')
k_142.plot(
    n=60,
    area=['V','KL'],
    width=1000,
    height=600,
    klines=['ma5', 'ma60', 'ma250'],
    is_notebook=True
)
```

可以看到，中国国贸的股价时而位于 5 日均线之上，时而位于 5 日均线之下，致使 5 日均线时而上行，时而下行。因此对于中国国贸这一段的走势而言，5 日均线实在是过于灵敏了，显得非常不稳定。

再看中国国贸的 250 日均线，6 月 21 日，中国国贸的 250 日均线在 14.43 元。22 日，其 250 日均线在 14.42 元。23 日，其 250 日均线在 14.42 元。24 日（周五），其 250 日均线在 14.41 元。27 日（周一），其 250 日均线在 14.42 元，250 日均线自此正式进入上行趋势。27 日的收盘价为 19.04 元，假设我们是以当日的收盘价买入。

8 月 15 日，中国国贸的股价见到最高 23.02 元，之后一路震荡下跌。12 月 12 日，中国国贸的 250日均线在 17.29 元，而当天的收盘价在 17.13 元，收盘价跌破了 250 日均线。13 日、14 日，其 250 日均线皆在 17.30 元，15 日至 26 日，其 250 日均线皆在 17.31 元，27 日、28 日，其 250 日均线皆在 17.30 元。29 日，其 250 日均线在 17.28 元。30日，其 250 日均线在 17.27 元可以看到，250 日均线是在 12 月 27 日正式进入下行趋势的，当天的收盘价为 16.09 元。假设我们是以当日的收盘价卖出，则将亏损 1 - 16.09 ÷ 19.04 = 15.49% 

在这个案例中，均线滞后性这一缺点展露无遗。我们根据 250 日均线上行买入是在 6 月 27 日 19.04 元，相较于股价见底时的 1 月 14 日的 11.91 元买入时间滞后了五个多月，买入价格高出了 59.87%。我们根据 250 日均线下行卖出是在 12 月的 16.09 元，相较于股价见顶时的 8 月 15 日的 23.02 元，卖出时间滞后了四个多月，卖出价格低了 30.10%。

由此可见，**要想通过均线来研判趋势，就必须深刻体悟到均线既稳定又滞后的性质**。唯其如此，才能够扬长避短，制订稳健可行的操作计划。若只是简单地将均线视为无往不利的神器，则极有可能不是因为选取的均线过于灵敏而导致频繁买卖，就是因为选取的均线过于滞后而导致买卖太迟。亦如赵藩在成都武侯祠所撰写的那副著名的“攻心联”。
